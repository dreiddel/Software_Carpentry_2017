```python
def celsius_fahrenheit(C):
    # F = (C * 9/5) + 32
    fahrenheit = (C * (9/5)) + 32
    print(fahrenheit)
    return fahrenheit
```

```python
boilingF = celsius_fahrenheit(100)
```

```python
print(boilingF)
```

```python
def fahrenheit_celsius(F=None):
    if F == None:
        print("Celsius Boiling Point = 100.")
        return
    # C = (F-32)*5/9
    celsius = (F-32)*5/9
    print(celsius)
    return celsius
```

```python
fahrenheit_celsius()
```

```python
import pandas as pd
```

```python
dem_health_data = pd.read_csv("Dem_Health_Full.txt",sep = '\t')
```

```python
dem_health_data.sort_values(by="Population_Density")
```

```python
def get_urban(df, threshold):
    urban = df.loc[df['Population_Density'] >= threshold, :]
    return urban
```

```python
urban_counties = get_urban(dem_health_data, 100)
urban_counties.head()
```

```python
urban_counties.sort_values(by = "Population_Density").tail(5)
```

```python
def split_county_by_pop(df, threshold):
    rural_county = df.loc[df['Population_Density'] <  threshold, :]
    urban_county = df.loc[df['Population_Density'] >= threshold, :]
    return [urban_county, rural_county]
```

```python
urban_rural_counties = split_county_by_pop(dem_health_data, 500)
```
